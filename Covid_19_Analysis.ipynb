```python
import pandas as pd 
import ipywidgets as widgets 
from IPython.display import display 
import matplotlib.pyplot as plt
```

# Data viz

To display real-time statistics OF COVID-19 concerning a specific region the user can select a region/contry from the list below to display


```python
data=pd.read_csv("covid.csv",sep=',',index_col=0)
```

```python
select_country=widgets.Dropdown(
    options=set(data['Country']),
    value='France',
    description='Region:',
    disabled=False
)
print("Chose a region or a country from the options below:")
display(select_country)
```

```python
data[data['Country']==select_country.value].plot(x='Date',y='Confirmed',title='Number of Confirmed cases in : %s '%(select_country.value))
```

```python
# gca stands for 'get current axis'
ax = plt.gca()
data[data['Country']==select_country.value].plot(x='Date',y='Confirmed',ax=ax)
data[data['Country']==select_country.value].plot(x='Date',y='Recovered', color='green', ax=ax)
data[data['Country']==select_country.value].plot(x='Date',y='Death', color='red', ax=ax)
ax.set_title('   Stats of Covid-19 in %s during the last few months'%(select_country.value))
plt.show()
```
